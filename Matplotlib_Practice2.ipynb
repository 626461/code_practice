## Plot Appearance


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,alpha=0.5,linestyle='-.',marker='o',markersize=10,markerfacecolor='green',markeredgewidth=5,markeredgecolor='red')
# RGB Hex Code
```

```python
fig.savefig('my_pic3.png', dpi=100)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,alpha=0.5,linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
fig.savefig('my_image.png',dpi=100)
```
