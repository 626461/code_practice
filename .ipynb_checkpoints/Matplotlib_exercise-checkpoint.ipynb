```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y,color='red')
axes2.plot(x,y,color='red')


axes1.set_title('Figure1')
plt.xlabel('X Label')
plt.ylabel('Y Label')

axes2.set_title('Figure2')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2 = fig.add_axes([0.2,0.5,.4,.4])
axes2.plot(x,y)
axes2.set_title('Zoom')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim(20,22)
axes2.set_ylim(30,50)
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
fig,axes = plt.subplots(1,2)
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y,color='blue',lw=3)
axes[1].plot(x,z,color='red',lw=3,ls='--')
```
