## Subplots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #

axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left, bottom, width & height

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(x,y)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)  # in subplot we can define no of rows and columns
axes =axes.flatten()
```

```python
fig, axes = plt.subplots(nrows=3,ncols=5)  # in subplot we can define no of rows and columns
axes = axes.flatten()
for ax in axes:
    ax.plot(x, y)

plt.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)   ## to save the picture in folder, DPI is dots-per-inch (pixel per inch)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_title('X_Label')
ax.set_title('Y_Label')
```

```python
fig.savefig('my_pic.png', dpi=100)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='Y Cubed')

ax.legend(loc=0)
```

```python
fig.savefig('my_pic2.png', dpi=100)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='Y Cubed')

ax.legend(loc=(0.1,0.1)) # for position
```

```python
fig.savefig('my_pic3.png', dpi=100)
```
