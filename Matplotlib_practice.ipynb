```python
plt.subplot(1,2,1)  #(row,column,plot no)
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot 2')
```

# Object oriented method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([.5,.5,.9,.9])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #

axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left, bottom, width & height

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(x,y)
axes2.set_title('Smaller Plot')
```
